```python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attrs):
        self.attrs = attrs
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        return X[self.attrs].values
```

```python
num_attrs = ["Pclass", "Age", "SibSp", "Parch", "Fare"]
cat_attrs = ["Sex", "Embarked"]
```

```python
sibsp_ix, parch_ix = 2, 3

class AttribAdder(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        famSize = X[:, sibsp_ix] + X[:, parch_ix] + 1
        return np.c_[X, famSize]
```

##### HANDLING MISSING NUMERICAL VALUES

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
    ("selector", DataFrameSelector(num_attrs)),
    ("imputer", SimpleImputer(strategy="median")),
    ("attrib_adder", AttribAdder())
])
```

##### HANDLING MISSING CATEGORICAL VALUES

```python
from sklearn.preprocessing import OneHotEncoder

cat_pipeline = Pipeline([
    ("selector", DataFrameSelector(cat_attrs)),
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("label_binarize", OneHotEncoder(sparse_output=False))
])
```

```python
from sklearn.pipeline import FeatureUnion

full_pipeline = FeatureUnion([
    ("num", num_pipeline),
    ("cat", cat_pipeline)
])

# Use fit transfrom on full pipeline
titanic_prepared = full_pipeline.fit_transform(df)
titanic_prepared_df = pd.DataFrame(titanic_prepared, columns=[
    "Pclass", "Age", "SibSp", "Parch", "Fare", "FamSize", "Female", "Male", "C", "Q", "S"
])
titanic_prepared_df
```

```python
labels = df['Survived']
```

```python
estimators = [SVC(), RandomForestClassifier(), KNeighborsClassifier(), SGDClassifier()]
fig, ax = plt.subplots(len(estimators), 2, figsize=(20, 20))

def run_compare(estis, cv=3):
    for esti_ix, esti in enumerate(estis):
        esti_preds = cross_val_predict(esti, titanic_prepared, labels, cv=cv)
        
        esti_pr_disp = PrecisionRecallDisplay.from_predictions(labels,
                                                               esti_preds, ax=ax[esti_ix][0], name=esti)
        esti_roc_disp = RocCurveDisplay.from_predictions(labels,
                                                         esti_preds, ax=ax[esti_ix][1], name=esti)
        
        print(f"\nClassification Report for {esti}:")
        print(classification_report(labels, esti_preds))

# Run
run_compare(estimators, cv=5)
plt.show()
```

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns
```

```python
model_performance = {
    'SVC': {'precision': 0.82, 'recall': 0.78, 'f1': 0.80},
    'RandomForest': {'precision': 0.85, 'recall': 0.81, 'f1': 0.83},
    'KNeighbors': {'precision': 0.79, 'recall': 0.75, 'f1': 0.77},
    'SGD': {'precision': 0.80, 'recall': 0.77, 'f1': 0.78}
}

# Create figure with subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 16))

# 1. Feature Correlation Network
def create_correlation_network(corr_matrix, threshold=0.3, ax=None):
    G = nx.Graph()
    
    # Add nodes
    for feature in corr_matrix.columns:
        G.add_node(feature)
    
    # Add edges for correlations above threshold
    for i, feature1 in enumerate(corr_matrix.columns):
        for j, feature2 in enumerate(corr_matrix.columns):
            if i < j:  # Avoid duplicate edges
                corr_val = abs(corr_matrix.iloc[i, j])
                if corr_val > threshold:
                    G.add_edge(feature1, feature2, weight=corr_val, 
                              color='red' if corr_matrix.iloc[i, j] < 0 else 'blue')
    
    # Create layout
    pos = nx.spring_layout(G, k=2, iterations=50)
    
    # Draw network
    nx.draw_networkx_nodes(G, pos, ax=ax, node_color='lightcoral', 
                          node_size=2000, alpha=0.8)
    nx.draw_networkx_labels(G, pos, ax=ax, font_size=10, font_weight='bold')
    
    # Draw edges with different colors for positive/negative correlations
    edges = G.edges()
    colors = [G[u][v]['color'] for u, v in edges]
    weights = [G[u][v]['weight'] * 5 for u, v in edges]  # Scale for visibility
    
    nx.draw_networkx_edges(G, pos, ax=ax, edge_color=colors, 
                          width=weights, alpha=0.6)
    
    ax.set_title('Feature Correlation Network\n(Red: Negative, Blue: Positive)', 
                fontsize=14, fontweight='bold')
    ax.axis('off')
    
    return G

# 2. Model Performance Comparison Network
def create_model_network(model_perf, ax=None):
    G = nx.Graph()
    
    # Add model nodes
    for model in model_perf.keys():
        G.add_node(model, node_type='model')
    
    # Add metric nodes
    metrics = ['precision', 'recall', 'f1']
    for metric in metrics:
        G.add_node(metric, node_type='metric')
    
    # Connect models to metrics with performance as edge weight
    for model, perf in model_perf.items():
        for metric, score in perf.items():
            G.add_edge(model, metric, weight=score)
    
    # Create layout
    pos = {}
    # Position models in a circle
    model_positions = nx.circular_layout(list(model_perf.keys()), scale=2)
    pos.update(model_positions)
    
    # Position metrics in the center
    metric_positions = nx.circular_layout(metrics, scale=0.5)
    pos.update(metric_positions)
    
    # Draw model nodes
    model_nodes = [n for n in G.nodes() if G.nodes[n].get('node_type') == 'model']
    nx.draw_networkx_nodes(G, pos, nodelist=model_nodes, ax=ax,
                          node_color='lightblue', node_size=2500, alpha=0.8)
    
    # Draw metric nodes
    metric_nodes = [n for n in G.nodes() if G.nodes[n].get('node_type') == 'metric']
    nx.draw_networkx_nodes(G, pos, nodelist=metric_nodes, ax=ax,
                          node_color='lightgreen', node_size=2000, alpha=0.8)
    
    # Draw edges with thickness based on performance
    for model in model_perf.keys():
        for metric, score in model_perf[model].items():
            nx.draw_networkx_edges(G, pos, edgelist=[(model, metric)], ax=ax,
                                 width=score*5, alpha=0.6, edge_color='gray')
    
    # Add labels
    nx.draw_networkx_labels(G, pos, ax=ax, font_size=9, font_weight='bold')
    
    ax.set_title('Model Performance Network\n(Edge thickness = Performance score)', 
                fontsize=14, fontweight='bold')
    ax.axis('off')

# 3. Hierarchical Feature Importance Tree
def create_feature_tree(ax=None):
    G = nx.DiGraph()
    
    # Add root node
    G.add_node("Titanic\nSurvival", level=0)
    
    # Add category nodes
    categories = ["Passenger\nClass", "Demographics", "Family\nStructure"]
    for cat in categories:
        G.add_node(cat, level=1)
        G.add_edge("Titanic\nSurvival", cat)
    
    # Add specific features
    feature_mapping = {
        "Passenger\nClass": ["Pclass", "Fare"],
        "Demographics": ["Age", "Gender"],
        "Family\nStructure": ["SibSp", "Parch", "FamSize"]
    }
    
    for cat, features in feature_mapping.items():
        for feature in features:
            G.add_node(feature, level=2)
            G.add_edge(cat, feature)
    
    # Create hierarchical layout
    pos = nx.nx_agraph.graphviz_layout(G, prog='dot')
    
    # Draw nodes by level
    level_colors = {0: 'red', 1: 'orange', 2: 'lightblue'}
    level_sizes = {0: 3000, 1: 2000, 2: 1500}
    
    for level in range(3):
        level_nodes = [n for n in G.nodes() if G.nodes[n].get('level', 2) == level]
        nx.draw_networkx_nodes(G, pos, nodelist=level_nodes, ax=ax,
                              node_color=level_colors[level], 
                              node_size=level_sizes[level], alpha=0.8)
    
    # Draw edges
    nx.draw_networkx_edges(G, pos, ax=ax, arrows=True, arrowsize=20,
                          edge_color='gray', width=2)
    
    # Add labels
    nx.draw_networkx_labels(G, pos, ax=ax, font_size=9, font_weight='bold')
    
    ax.set_title('Feature Hierarchy Tree', fontsize=14, fontweight='bold')
    ax.axis('off')

# 4. Passenger Survival Network (Sample)
def create_survival_network(ax=None):
    G = nx.Graph()
    
    # Sample passenger data - categories based on survival factors
    passenger_groups = {
        'First_Class_Women': {'survived': 0.97, 'size': 94},
        'First_Class_Men': {'survived': 0.37, 'size': 122},
        'Second_Class_Women': {'survived': 0.92, 'size': 76},
        'Second_Class_Men': {'survived': 0.16, 'size': 108},
        'Third_Class_Women': {'survived': 0.50, 'size': 144},
        'Third_Class_Men': {'survived': 0.14, 'size': 347}
    }
    
    # Add nodes
    for group, data in passenger_groups.items():
        G.add_node(group, survival_rate=data['survived'], 
                  size=data['size'], survived=data['survived'] > 0.5)
    
    # Add edges between similar groups
    edges = [
        ('First_Class_Women', 'First_Class_Men'),
        ('Second_Class_Women', 'Second_Class_Men'),
        ('Third_Class_Women', 'Third_Class_Men'),
        ('First_Class_Women', 'Second_Class_Women'),
        ('Second_Class_Women', 'Third_Class_Women'),
        ('First_Class_Men', 'Second_Class_Men'),
        ('Second_Class_Men', 'Third_Class_Men')
    ]
    
    G.add_edges_from(edges)
    
    # Create layout
    pos = nx.spring_layout(G, k=3, iterations=50)
    
    # Color nodes based on survival rate
    node_colors = []
    node_sizes = []
    for node in G.nodes():
        survival_rate = G.nodes[node]['survival_rate']
        node_colors.append(plt.cm.RdYlGn(survival_rate))
        node_sizes.append(G.nodes[node]['size'] * 10)
    
    # Draw network
    nx.draw_networkx_nodes(G, pos, ax=ax, node_color=node_colors,
                          node_size=node_sizes, alpha=0.8)
    nx.draw_networkx_edges(G, pos, ax=ax, edge_color='gray', alpha=0.5)
    
    # Add labels
    labels = {node: node.replace('_', '\n') for node in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels, ax=ax, font_size=8, font_weight='bold')
    
    ax.set_title('Passenger Group Survival Network\n(Color: Survival Rate, Size: Group Size)', 
                fontsize=14, fontweight='bold')
    ax.axis('off')

# Create all diagrams
create_correlation_network(corr_df, threshold=0.2, ax=ax1)
create_model_network(model_performance, ax=ax2)
create_feature_tree(ax=ax3)
create_survival_network(ax=ax4)

plt.tight_layout()
plt.show()

# Print network statistics
print("Network Analysis Summary:")
print("========================")
print("1. Feature Correlation Network:")
print("   - Shows relationships between features")
print("   - Red edges: Negative correlations")
print("   - Blue edges: Positive correlations")
print("   - Thickness: Correlation strength")
print()
print("2. Model Performance Network:")
print("   - Central nodes: Evaluation metrics")
print("   - Outer nodes: ML models")
print("   - Edge thickness: Performance score")
print()
print("3. Feature Hierarchy Tree:")
print("   - Root: Survival prediction")
print("   - Branches: Feature categories")
print("   - Leaves: Individual features")
print()
print("4. Passenger Survival Network:")
print("   - Nodes: Passenger groups")
print("   - Colors: Survival rates (green=high, red=low)")
print("   - Size: Group size")
```
