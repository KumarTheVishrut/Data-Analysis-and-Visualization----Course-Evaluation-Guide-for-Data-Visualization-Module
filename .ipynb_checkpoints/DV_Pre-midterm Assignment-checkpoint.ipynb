### Cleaning the dataset

```python
print(f"Dataset shape: {df.shape}")
print(f"Total records: {len(df)}")
```

```python
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df)) * 100
    
missing_data
```

```python
missing_percent
```

```python
# Changing the date to datetime for formatting
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# making a copy of the clean Data frame
df_clean = df.copy()
```

```python
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
df_clean['Time'] = pd.to_datetime(df_clean['Time'], format='%H:%M:%S').dt.time
  
```

```python
# creating a separate coluimn for date and time concatenated
df_clean['DateTime'] = pd.to_datetime(df_clean['Date'].astype(str) + ' ' + df_clean['Time'].astype(str))
```

```python
df_clean['Hour'] = df_clean['DateTime'].dt.hour
df_clean['DayOfWeek'] = df_clean['DateTime'].dt.day_name()
df_clean['Month'] = df_clean['DateTime'].dt.month
```

```python
df_clean
```

```python
status_counts = df_clean['Booking Status'].value_counts()
status_counts
```

```python
print(f"Percentage:\n{(status_counts/len(df_clean)*100).round(2)}")
```

```python
# creating a mapping to handle missing values
mapping = {
    'Cancelled by Customer': 'Customer Cancelled',
    'Cancelled by Driver': 'Driver Cancelled',
    'Incomplete': 'Incomplete',
    'Completed': 'Completed',
    'No Driver Found': 'Driver Unavailable'
}

df_clean['Ride_Type'] = df_clean['Booking Status'].map(mapping).fillna('Unknown')
```

```python
df_clean
```

```python
completed_mask = df_clean['Ride_Type'] == 'Completed'
```

```python
completed_mask
```

```python
completed_rides = df_clean[completed_mask].copy()
```

```python
# dataset comprising of completed rides
completed_rides
```

```python
missing_in_completed = completed_rides.isnull().sum()
print(missing_in_completed[missing_in_completed > 0])
```

```python
# filling missing driver ratings
if 'Driver Ratings' in completed_rides.columns:
        median_driver_rating = completed_rides['Driver Ratings'].median()
        median_customer_rating = completed_rides['Customer Rating'].median()
        
        print(f"\nMedian Driver Rating: {median_driver_rating}")
        print(f"Median Customer Rating: {median_customer_rating}")
```

```python
# Filling the missing ratings
completed_rides['Driver Ratings'].fillna(median_driver_rating, inplace=True)
completed_rides['Customer Rating'].fillna(median_customer_rating, inplace=True)
```

```python
df_clean.loc[completed_mask, :] = completed_rides
```

```python
df_clean
```

```python
# Handling cancelled rides

customer_cancelled = df_clean['Ride_Type'] == 'Customer Cancelled'
driver_cancelled = df_clean['Ride_Type'] == 'Driver Cancelled'
```

```python
print(f"Customer cancelled trips: {customer_cancelled.sum()}")
print(f"Driver cancelled trips: {driver_cancelled.sum()}")
```

```python
 cancelled_rides = df_clean[customer_cancelled | driver_cancelled].copy()
```

```python
cancelled_rides
```

```python
financial_cols = ['Booking Value', 'Ride Distance']
rating_cols = ['Driver Ratings', 'Customer Rating']
    
for col in financial_cols:
    if col in cancelled_rides.columns:
        cancelled_rides[col].fillna(0, inplace=True)
    
for col in rating_cols:
    if col in cancelled_rides.columns:
        cancelled_rides[col].fillna(0, inplace=True)
```

```python
df_clean.loc[customer_cancelled | driver_cancelled, :] = cancelled_rides
```

```python
df_clean
```

```python
# handling incomplete rides

incomplete_mask = df_clean['Ride_Type'] == 'Incomplete'
incomplete_rides = df_clean[incomplete_mask].copy()
```

```python
print(f"Incomplete rides count: {incomplete_mask.sum()}")
```

```python
rating_cols = ['Driver Ratings', 'Customer Rating']
for col in rating_cols:
    if col in incomplete_rides.columns:
        incomplete_rides[col].fillna(1.0, inplace=True)
```

```python
if 'Booking Value' in incomplete_rides.columns:
        incomplete_rides['Booking Value'].fillna(0, inplace=True)
```

```python
df_clean.loc[incomplete_mask, :] = incomplete_rides
```

```python
df_clean
```

```python
remaining_missing = df_clean.isnull().sum()
print("Remaining missing values:")
print(remaining_missing[remaining_missing > 0])
```

```python
# cleaning numeric columns of a dataframe 

num_cols = ['Avg VTAT', 'Avg CTAT', 'Booking Value', 'Ride Distance', 
            'Driver Ratings', 'Customer Rating']

for col in num_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())
```

```python
remaining_missing = df_clean.isnull().sum()
print("Remaining missing values:")
print(remaining_missing[remaining_missing > 0])
```

## Simple visualizations and statistics to get contextual enquiry of the data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
num_cols = df_clean.select_dtypes(include='number').columns.tolist()

# Plot first 6 numeric columns
for c in num_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df_clean[c], bins=40, kde=True, color="magenta")  # kde=True gives smooth curve like Plotly
    plt.title(f'Histogram: {c}', fontsize=14)
    plt.xlabel(c)
    plt.ylabel("Count")
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
rides_per_day = df_clean.groupby(df_clean['DateTime'].dt.date).size()

# Plot
plt.figure(figsize=(14,6))
plt.plot(rides_per_day.index, rides_per_day.values, marker='*', linestyle='-', color = 'magenta')
plt.title("Daily Ride Volume", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Rides", fontsize=12)
plt.grid(True, alpha=0.5)
plt.show()
```

```python
day_counts = df_clean['DayOfWeek'].value_counts()
day_counts
```

```python
plt.figure(figsize=(8,5))
plt.plot(day_counts.index, day_counts.values, marker='o', color='magenta', linewidth=2)

for x,y in zip(day_counts.index, day_counts.values):
    plt.text(x, y+200, str(y), ha='center')

plt.title("Rides by Day of the Week", fontsize=16)
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Number of Rides", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
payment_counts = df_clean['Payment Method'].value_counts()

cmap = plt.cm.viridis
colors = cmap(np.linspace(0, 1, len(payment_counts)))

plt.figure(figsize=(8,6))
wedges, texts, autotexts = plt.pie(
    payment_counts,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops={'fontsize': 10}
)

plt.legend(wedges, payment_counts.index, title="Payment Method", loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Payment Method Distribution", fontsize=16)

plt.show()
```

```python
vehicle_counts = df_clean['Vehicle Type'].value_counts()

cmap = plt.cm.viridis
colors = cmap(np.linspace(0, 1, len(payment_counts)))

plt.figure(figsize=(8,6))
wedges, texts, autotexts = plt.pie(
    vehicle_counts,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops={'fontsize': 10}
)

plt.legend(wedges, vehicle_counts.index, title="Vehicle Mode", loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("Vehicle mode distribution", fontsize=16)

plt.show()
```

```python
loc_matrix = pd.crosstab(df_clean["Pickup Location"], df_clean["Drop Location"])
matrix_top = loc_matrix.iloc[:15, :15]

mask = np.triu(np.ones_like(matrix_top, dtype=bool))
plt.figure(figsize=(12,8))
sns.heatmap(matrix_top, cmap="YlGnBu", mask=mask, annot=False, linewidths=0.5)
plt.title("Pickup vs Drop Location Heatmap (Top 15x15 - Lower Triangle Only)")
plt.show()
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
locations = df_clean['Pickup Location'].unique()

geolocator = Nominatim(user_agent="pickup_mapper")

coords = []
for loc in locations:
    try:
        location = geolocator.geocode(f"{loc}, Delhi, India")
        if location:
            coords.append((loc, location.latitude, location.longitude))
    except:
        coords.append((loc, None, None))

df_coords = pd.DataFrame(coords, columns=['Pickup Location', 'lat', 'lon'])
df_coords = df_coords.dropna()

# convert to GeoDataFrame
gdf = gpd.GeoDataFrame(df_coords, geometry=gpd.points_from_xy(df_coords.lon, df_coords.lat), crs="EPSG:4326")

# plot on basemap
gdf = gdf.to_crs(epsg=3857)
ax = gdf.plot(figsize=(10, 10), color="red", markersize=50, alpha=0.7)
ctx = __import__("contextily")
ctx.add_basemap(ax, source=ctx.providers.CartoPositron)
plt.title("Pickup Locations in Delhi")
plt.show()
```
